```python
import os
import music_tag
```

```python
all_songs = sorted(os.listdir('All songs'))
song_dir = 'All songs/'
image_dir = 'All images/'
```

```python
all_data = []
for song in all_songs:
    temp = []
    audio_file = music_tag.load_file(song_dir + song)

    temp.append(song)
    temp.append(str(audio_file['tracktitle']))
    temp.append(str(audio_file['artist']))
    temp.append(str(audio_file['album']))
    temp.append(int(audio_file['tracknumber']))
    temp.append(str(audio_file['albumartist']))
    temp.append(int(audio_file['year']))
    all_data.append(temp)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(all_data)
df.columns = ['audio file', 'title', 'artist', 'album', 'track no', 'album artist', 'year']
df.sort_values('title')
```
