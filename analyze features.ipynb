```python
df = pd.read_csv('song_dataset_discogs.csv')
df.sort_values('song name', inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
genre_columns, tag_columns, theme_columns, instrument_columns, nsynth_columns, others = ['song name'], ['song name'], ['song name'], ['song name'], ['song name'], ['song name']
for col in df.columns:
    if col[:5] == "Genre":
        genre_columns.append(col)
    elif col[:3] == "Tag":
        tag_columns.append(col)
    elif col[:5] == "Theme":
        theme_columns.append(col)
    elif col[:10] == "Instrument":
        instrument_columns.append(col)
    elif col[:6] == "Nsynth":
        nsynth_columns.append(col)
    else:
        others.append(col)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
minmax = MinMaxScaler(feature_range=(0,1))
X = df[instrument_columns].drop(columns='song name')
X = minmax.fit_transform(X)

df_cosine=pd.DataFrame(cosine_similarity(X,dense_output=True))
indices = pd.Series(df.index, index = df['song name'])
```

```python
def similar_songs(name, similarity = df_cosine):
    index = indices[name]
    similarity_scores = list(enumerate(similarity[index]))
    similarity_scores = sorted(similarity_scores, key = lambda x: x[1],reverse = True)
    similarity_scores = similarity_scores[1:6]
    res_indices = [i[0] for i in similarity_scores]
    return df['song name'].iloc[res_indices[:]]
```

```python
top5_songs = similar_songs('Kesariya - Pritam, Arijit Singh, Amitabh Bhattacharya', df_cosine)
for song in top5_songs:
    print(song)
```
